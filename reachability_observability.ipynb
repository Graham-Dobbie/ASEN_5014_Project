```python
import numpy as np
import scipy as sci
import control
```

```python
## Define SS

RE = 6378 # km
ROrb = 300 # km
r0 = RE + ROrb
mu = 398600


A = np.array([[           0, 1, 0,                0],
                [3*mu/(r0**3), 0, 0, 2*np.sqrt(mu/r0)],
                [ 0, 0, 0, 1],
                [0, -2*np.sqrt(mu/r0**5), 0, 0 ]])

B = np.array([[0,0],
                [1,0],
                [0,0],
                [0,1/r0]])

C = np.array([[1,0,0,0],[0,0,1,0]])

D = np.zeros([2,2])
```

```python
## Open loop poles

Evals, V = np.linalg.eig(A)

print(f"Open loop poles: \n l_1 = {Evals[0]} \n l_2 = {Evals[1]} \n l_3 = {Evals[2]} \n l_4 = {Evals[3]}")
```

```python
## Controllability Matrix

P = control.ctrb(A,B)
print(np.linalg.matrix_rank(P))
print(f"P = {P}")
O = control.obsv(A,C)
print(f"O = {O}")
print(np.linalg.matrix_rank(O))
```
